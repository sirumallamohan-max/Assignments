### Exercise 1: Prime Numbers
A **prime number** is a natural number greater than 1 that has no positive divisors other than 1 and itself, 
2 is the only positive  number which is prime number

```python
# to check a given number is prime or not
# take input and store in a  variable
number =  int(input("Enter a number :")) # take user input and convert to int
#number =  6 # directly assign and check
if number <=1 : # to check 1 or below
    print (f"{number} is not a prime number")     # f for formating
elif number== 2: # to check for 2
    print(f"{number} is a prime number")
elif number % 2 ==0: # check even number
    print (f"{number} is not a prime number")
else:
    check =True # initial value True boolen
    for i in range(3,number,2):
        if number % i == 0:
            check=False
            break
    
    if check==True:
       print(f"{number} is a prime number")
    else:
       print(f"{number} is not a prime number")

# end of the code
```

An **Optimized Approach**  in the above ,we are checking upto n number. We can check upto Square Root of n number

```python
import math

# Optimized Approach : to check a given number is prime or not 
# Need to check upto Square root of n
number =  int(input("Enter a number :")) # take user input and convert to int
#number =  6 # directly assign and check
if number <=1 : # to check 1 or below
    print (f"{number} is not a prime number")     # f for formating
elif number== 2: # to check for 2
    print(f"{number} is a prime number")
elif number % 2 ==0: # check even number
    print (f"{number} is not a prime number")
else:
    check =True # initial value True boolen
    for i in range(3,int(math.sqrt(number))+1): # here we are using sqrt function
        if number % i == 0:
            check=False
            break
    
    if check==True:
       print(f"{number} is a prime number")
    else:
       print(f"{number} is not a prime number")

# end of the code
```
