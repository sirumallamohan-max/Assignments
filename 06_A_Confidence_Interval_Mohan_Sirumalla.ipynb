**Background**\
In quality control processes, especially when dealing with high-value items, destructive sampling is a necessary but costly method to ensure product quality. The test to determine whether an item meets the quality standards destroys the item, leading to the requirement of small sample sizes due to cost constraints.\
**Scenario**\
A manufacturer of print-heads for personal computers is interested in estimating the mean durability of their print-heads in terms of the number of characters printed before failure. To assess this, the manufacturer conducts a study on a small sample of print-heads due to the destructive nature of the testing process.\
**Data**\
A total of 15 print-heads were randomly selected and tested until failure. The durability of each print-head (in millions of characters) was recorded as follows:\
**1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29**


```python
## Wrire Code Here
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#data
data =[1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
# Check how the distibution is
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, color='skyblue', bins=len(set(data)) // 2 or 5, label='Distribution')
```

## Assignment Tasks

### a. Build 99% Confidence Interval Using Sample Standard Deviation


Assuming the sample is representative of the population, construct a 99% confidence interval for the mean number of characters printed before the print-head fails using the sample standard deviation.\
Explain the steps you take and the rationale behind using the t-distribution for this task.

**Confidence Interval Formula**\
A confidence interval (CI)=  Sample Mean ± (Critical Value × Standard Error),
## $$ CI = \bar{x} \pm t \frac {s}{\sqrt{n}}$$
|        |      |
|--------|------|
| CI = confidence Interval| s = sample standard deviation |
| $\bar{x}$= sample Mean  | n = sample size  |
|t = critical value |   |



```python
#Steps to Calculate a Confidence Interval
#1. Find Your Sample Statistics
#* Sample Size
#* Sample Mean
#* Sample Standard Deviation 
sample_size =len(data)
sample_mean =np.mean(data)
sample_std = np.std(data,ddof=1) # Delta degree of freedom number of elements -1
print("Sample Statistics :")
print(f"Mean : {sample_mean}, Sample: {sample_size}: Standard Deviation :{sample_std}")

confidence_level =99 # given
significance_level=1-confidence_level/100
degree_of_freedom =  sample_size -1
critical_value = 2.977  # from t-table based on confidence_level and degree of freedom
print("\nConfidence Level :")
print(f"Degree of Freedom (Sample -1): {degree_of_freedom}")
print(f"Confidence Level: {confidence_level}%, significance level {significance_level}")
print(f"Critical Value: {critical_value}")

#Calulcate the Standard Error (SE)
std_error = sample_std/ np.sqrt(sample_size)
print (f"\nStandard Error : {std_error}")

#Calculate the Margin of Errr (ME)
#   ME = Critival Value X SE
margin_of_error= critical_value * std_error
print (f"\nMargin of Error : {margin_of_error}")
# Construct the Interval
#  Lower Bound :mean -ME
#  upper bBoudn :mean + ME
lower_bound= sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error
print (f"\nUpper Bound : {lower_bound} , Lower Bound {upper_bound}")

print(f"\nSample Mean: {sample_mean}")
print(f"{confidence_level}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

### Explain the steps you take and the rationale behind using the t-distribution for this task.

**The t-distribution is used instead of the standard normal (Z) distribution for the following reasons:**\
**Unknown Population Standard Deviation**: The true population standard deviation ($\sigma$) is unknown, so we must estimate it using the sample standard deviation (s).\
**Small Sample Size**: The sample size (n=15) is less than the generally accepted threshold of 30 for safely approximating with the normal distribution when($\sigma$) is unknown.

### b. Build 99% Confidence Interval Using Known Population Standard Deviation
If it were known that the population standard deviation is 0.2 million characters, construct a 99% confidence interval for the mean number of characters printed before failure.
