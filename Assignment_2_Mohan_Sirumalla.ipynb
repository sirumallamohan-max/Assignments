1.	Create a list of 5 integers. Perform and print the result of the following operations: **append, extend, insert, remove, pop, clear, index, count, sort, and reverse**.


```python
# List of 5 integers
list_int = [15,12,25,43,21] # store the integers in a list
print("Original List",list_int)
# --> append() adds element at the end of the list
list_int.append(10)
print("After append " ,list_int)
# --> extend() adds mutliple elements at the end of the list
list_int.extend([20,10])
print("After extend " ,list_int)
# --> insert() adds element at a specific position
list_int.insert(2,30)
print("After insert " ,list_int)
# --> remove() removes the first occurrence of a element
list_int.remove(10)
print("After remove " ,list_int)
# --> pop() removes the last element
list_int.pop()
print("After pop    " ,list_int)
# --> clear() removes all the items
list_int.clear()
print("After clear  " ,list_int)
# --> index() searched
#list_int.index(0) # it will given error list is empty now
# to continue need to add items in the list
list_int.extend([50, 40, 10, 20, 30]) 
list_int.index(20)
# --> count() of particular item in the list

print("Count of 12 in the list =" ,list_int.count(12) )
# --> reverse
print("Before reverse",list_int)
list_int.reverse()
print("After reverse ",list_int)
# --> sort the items in the order
list_int.sort()
print("after sort    ", list_int)
# --> sort the items in the order
list_int.sort(reverse=True)
print("after sort descending    ", list_int)

```
